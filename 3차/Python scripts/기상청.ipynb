```python
area2 = area.str.replace('_x000D_', '')
area2 = area2.str.replace('\t', '')
area2 = area2.str.replace('\xa0', '')
area2 = area2.str.split("\n")
```

```python
df_area = pd.DataFrame(area2.to_list(), columns = ['지역1', '지역2'])
```

```python
df_area['지역2'].nunique()
```

```python
area_dict = {"경상북도" : "경상도", "경상남도" : "경상도", "전라북도" : "전라도", "전라남도" : "전라도",
             "충청북도" : "충청도", "충청남도" : "충청도"}
```

```python
def area_map(x):
    try:
        return area_dict[x]
    except:
        return x
```

```python
df_area['지역1_2'] = df_area['지역1'].apply(area_map)
```

```python
point_loc = pd.read_excel("./지점_위치.xlsx")
```

```python
def search_point(x):
    best = None
    best_point = None

    for i, v in enumerate(point_loc.index):
        tmp_point = (point_loc.loc[v, '위도'], point_loc.loc[v, '경도'])
        hav = haversine(x, tmp_point, unit = 'km')
        if not best:
            best = hav
            best_point = point_loc.loc[v, '지점번호']
        else:
            if best > hav:
                best = hav
                best_point = point_loc.loc[v, '지점번호']

    return best_point
```

```python
search_point((36.5040736, 127.2494855))
```

```python
for i in tmp123.index:
    if i % 100 == 0:
        print(i)
    tmp123.loc[i, 'point'] = search_point((tmp123.loc[i, 'latitude'], tmp123.loc[i, 'longitude']))
```

```python
tmp123.to_csv("./위치별_관측소.csv", index = False)
```
